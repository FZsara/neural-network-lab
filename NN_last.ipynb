```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Data
data = np.array([[15], [16], [17], [18], [20], [21], [22], [25], [26], [27], [30], [32], [33], [34], [36], [37], [38], [40], [42], [45]])

# Number of clusters
k = 3

# Standardizing the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Applying KMeans
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(data_scaled)

# Extracting cluster labels and centroids
labels = kmeans.labels_
centroids = scaler.inverse_transform(kmeans.cluster_centers_)

# Plotting the results
plt.figure(figsize=(8, 6))
for i in range(k):
    cluster_points = data[labels == i]
    plt.scatter(cluster_points, np.zeros_like(cluster_points), label=f'Cluster {i+1}')

plt.scatter(
    centroids, [0] * len(centroids),
    s=200, c='red', marker='X', label='Centroids'
)

plt.title('K-Means Clustering')
plt.xlabel('Data Values')
plt.yticks([])
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Data
data = np.array([[15], [16], [17], [18], [20], [21], [22], [25], [26], [27], [30], [32], [33], [34], [36], [37], [38], [40], [42], [45]])

# Number of clusters
k = 3

# Standardizing the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Applying KMeans
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(data_scaled)

# Extracting cluster labels and centroids
labels = kmeans.labels_
centroids_scaled = kmeans.cluster_centers_
centroids = scaler.inverse_transform(centroids_scaled)

# Printing the clusters and their values
for i in range(k):
    cluster_points = data[labels == i]
    print(f"Cluster {i + 1}: {cluster_points.flatten()}")

# Printing the centroids
print("\nCentroids of the clusters:")
for i, centroid in enumerate(centroids):
    print(f"Centroid {i + 1}: {centroid[0]:.2f}")

# Plotting the results
plt.figure(figsize=(8, 6))
for i in range(k):
    cluster_points = data[labels == i]
    plt.scatter(cluster_points, np.zeros_like(cluster_points), label=f'Cluster {i+1}')

plt.scatter(
    centroids, [0] * len(centroids),
    s=200, c='red', marker='X', label='Centroids'
)

plt.title('K-Means Clustering')
plt.xlabel('Data Values')
plt.yticks([])
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Defining the overlapping triangular membership functions
def mu_low(x):
    if 15 <= x <= 17.2:
        return (x - 15) / (17.2 - 15)
    elif 17.2 <= x <= 21:
        return (21 - x) / (21 - 17.2)
    else:
        return 0

def mu_medium(x):
    if 20 <= x <= 25.2:
        return (x - 20) / (25.2 - 20)
    elif 25.2 <= x <= 32:
        return (32 - x) / (32 - 25.2)
    else:
        return 0

def mu_high(x):
    if 30 <= x <= 37.4:
        return (x - 30) / (37.4 - 30)
    elif 37.4 <= x <= 45:
        return (45 - x) / (45 - 37.4)
    else:
        return 0

# Generating data for visualization
x_values = np.linspace(15, 45, 1000)
low_values = [mu_low(x) for x in x_values]
medium_values = [mu_medium(x) for x in x_values]
high_values = [mu_high(x) for x in x_values]

# Define fuzzy rules for Mamdani approach
# Rule 1: If Temperature is Low, then Output is Low
# Rule 2: If Temperature is Medium, then Output is Medium
# Rule 3: If Temperature is High, then Output is High
def fuzzify(input_value):
    return mu_low(input_value), mu_medium(input_value), mu_high(input_value)

# Defining the centroid method for defuzzification (Mamdani)
def defuzzify(x_values, low, medium, high):
    numerator = np.sum(x_values * low) + np.sum(x_values * medium) + np.sum(x_values * high)
    denominator = np.sum(low) + np.sum(medium) + np.sum(high)
    return numerator / denominator if denominator != 0 else 0

# Applying Mamdani Inference System to find the output for the given range
low_membership = np.array(low_values)
medium_membership = np.array(medium_values)
high_membership = np.array(high_values)

# Defuzzify to get the crisp output
crisp_output = defuzzify(x_values, low_membership, medium_membership, high_membership)

# Plotting the overlapping membership functions
plt.figure(figsize=(10, 6))

plt.plot(x_values, low_values, label="Low Temperature", color="blue")
plt.plot(x_values, medium_values, label="Medium Temperature", color="green")
plt.plot(x_values, high_values, label="High Temperature", color="red")

# Adding annotations and labels
plt.title("Overlapping Triangular Membership Functions for Temperature (Mamdani Approach)")
plt.xlabel("Temperature (°C)")
plt.ylabel("Membership Degree")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(15, color='black', linewidth=0.5)
plt.axvline(45, color='black', linewidth=0.5)
plt.grid(alpha=0.3)
plt.legend()
plt.show()

# Output the result of the Mamdani approach
print("Crisp Output after applying Mamdani approach:", crisp_output)
```
